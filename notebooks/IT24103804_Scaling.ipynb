```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
df=pd.read_csv('IT24100391Feature_Engineering_Part01.csv')
```

```python
print(df.head())
```

```python
print("Shape of dataset:", df.shape)
```

```python
print(df.info())
```

```python
standard_cols = ['test_score', 'gpa_or_score']   
```

```python
print("Before scaling:")
print(df[standard_cols ].head())
```

```python
scaler_standard = StandardScaler()
df[standard_cols] = scaler_standard.fit_transform(df[standard_cols])
```

```python
print("\n After scaling:")
print(df[standard_cols ].head())
```

```python
minmax_cols   = ['study_duration','starting_salary_usd']
```

```python
print("Before scaling:")
before_scale=(df[minmax_cols ])
print(before_scale.head())
```

```python
scaler_minmax = MinMaxScaler()
```

```python
df[minmax_cols] = scaler_minmax.fit_transform(df[minmax_cols])
```

```python
print("After scaling:")
after_scale=df[minmax_cols ]
print(after_scale.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fig, axes = plt.subplots(2, 2, figsize=(14,10))

# Study Duration
sns.histplot(before_scale['study_duration'], kde=True, ax=axes[0,0], color="green")
axes[0,0].set_title("Study Duration - Before Scaling")

sns.histplot(after_scale['study_duration'], kde=True, ax=axes[0,1], color="red")
axes[0,1].set_title("Study Duration - After Scaling")

# Starting Salary
sns.histplot(before_scale['starting_salary_usd'], kde=True, ax=axes[1,0], color="green")
axes[1,0].set_title("Starting Salary - Before Scaling")

sns.histplot(after_scale['starting_salary_usd'], kde=True, ax=axes[1,1], color="red")
axes[1,1].set_title("Starting Salary - After Scaling")

plt.tight_layout()
plt.show()
```
